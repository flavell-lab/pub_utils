```python
import pandas as pd
import pub_utils as pu
```

## Monoamine Connectome

```python
path_csv = '../data/RipollSanchez2023/monoamine_connectome_08062023.csv'
title = 'Monoamine Connectome - RipollSanchez2023'
colorbar_label = '# Unique Ligand Projections\n(different receptors for same ligand is counted once)'
path_df = '../connectomes/preassembled/monoamine_RipollSanchez2023.csv'
path_fig = '../plots/monoamine_RipollSanchez2023.png'

# Read from csv
df = pd.read_csv(path_csv)

# Standardize and save
std_df = pu.standardize_dataframe(df, pu.AllHermNeurons)
std_df.to_csv(path_df)

# Plot and save
std_fig = pu.plot_connectome_matrix(std_df, title=title, colorbar_label=colorbar_label)
std_fig.savefig(path_fig, dpi=600, bbox_inches='tight')
```
