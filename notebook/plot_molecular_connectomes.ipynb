```python
import pandas as pd
import pub_utils as pu
```

## Monoamine Connectome

```python
# Read
monoamine_df = pd.read_csv('../data/RipollSanchez2023/monoamine_connectome_08062023.csv')

# Plot
monoamine_plot_df, monoamine_fig = pu.plot_connectome_matrix(monoamine_df, title='Monoamine Connectome')

# Save
savepath = '../plots/monoamine_connectome.png'
monoamine_fig.savefig(savepath, dpi=600, bbox_inches='tight')
```

## Neuropeptide Connectome

```python
# Read
npp_short_df = pd.read_csv('../data/RipollSanchez2023/NPP_connectome_short_range_01022024.csv')

# Plot
npp_short_plot_df, npp_short_fig = pu.plot_connectome_matrix(npp_short_df, title='Neuropeptide Connectome (Short Range)')

# Save
savepath = '../plots/NPP_connectome_short_range.png'
npp_short_fig.savefig(savepath, dpi=600, bbox_inches='tight')
```

```python
# Read
npp_mid_df = pd.read_csv('../data/RipollSanchez2023/NPP_connectome_mid_range_01022024.csv')

# Plot
npp_mid_plot_df, npp_mid_fig = pu.plot_connectome_matrix(npp_mid_df, title='Neuropeptide Connectome (Mid Range)')

# Save
savepath = '../plots/NPP_connectome_mid_range.png'
npp_mid_fig.savefig(savepath, dpi=600, bbox_inches='tight')
```

```python
# Read
npp_long_df = pd.read_csv('../data/RipollSanchez2023/NPP_connectome_long_range_01022024.csv')

# Plot
npp_long_plot_df, npp_long_fig = pu.plot_connectome_matrix(npp_long_df, title='Neuropeptide Connectome (Long Range)')

# Save
savepath = '../plots/NPP_connectome_long_range.png'
npp_long_fig.savefig(savepath, dpi=600, bbox_inches='tight')
```

## Save dataframes to file

```python
monoamine_plot_df.to_pickle('../processed/monoamine_connectome.pkl')
npp_short_plot_df.to_pickle('../processed/NPP_connectome_short_range.pkl')
npp_mid_plot_df.to_pickle('../processed/NPP_connectome_mid_range.pkl')
npp_long_plot_df.to_pickle('../processed/NPP_connectome_long_range.pkl')
```
