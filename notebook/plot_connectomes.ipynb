```python
import pandas as pd
import pub_utils as pu
```

## Monoamine Connectome

```python
# Read from csv
monoamine_df = pd.read_csv('../data/RipollSanchez2023/monoamine_connectome_08062023.csv')

# Standardize and save
std_df = pu.standardize_dataframe(monoamine_df, pu.AllHermNeurons)
std_df.to_pickle('../processed/monoamine_connectome.pkl')

# Plot and save
monoamine_fig = pu.plot_connectome_matrix(std_df, title='Monoamine Connectome')
monoamine_fig.savefig('../plots/monoamine_connectome.png', dpi=600, bbox_inches='tight')
```

## Individual NPP-GPCR pairs

```python
# Define
npp = 'FLP-10'
gpcr = 'DMSR-1'
file_idx = pu.get_file_for_pair(npp, gpcr)

# Read
ind_npp_df = pd.read_csv(f'../data/RipollSanchez2023/individual/individual_mid_range/{file_idx}.csv')

# Standardize and save
std_df = pu.standardize_dataframe(ind_npp_df, pu.AllHermNeurons)
std_df.to_pickle(f'../processed/NPP_connectome_mid_range_{npp}_{gpcr}.pkl')

# Plot and save
ind_npp_fig = pu.plot_connectome_matrix(std_df, title=f'{npp} -> {gpcr} Connectome')
savepath = f'../plots/NPP_connectome_mid_range_{npp}_{gpcr}.png'
ind_npp_fig.savefig(savepath, dpi=600, bbox_inches='tight')
```
